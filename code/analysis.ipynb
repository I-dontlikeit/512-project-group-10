```python
import pandas as pd
```

```python
df= pd.read_csv('../data/clean_data/clean_gender_inequality_index.csv')
```

```python
df.head()
```

```python
#remove .. from the data
df = df.replace('..', 0)
df = df.replace('Not known', 0)
```

```python

# Convert the 'HDI Rank' column to numeric
df['HDI Rank'] = pd.to_numeric(df['HDI Rank'], errors='coerce')

# Convert the 'Gender Inequality Index Value' column to numeric
df['Gender Inequality Index Value'] = pd.to_numeric(df['Gender Inequality Index Value'], errors='coerce')

# Convert the 'Gender Inequality Index Rank' column to numeric
df['Gender Inequality Index Rank'] = pd.to_numeric(df['Gender Inequality Index Rank'], errors='coerce')

# Convert the 'Women Share of Seats in Parliament' column to numeric
df['Women Share of Seats in Parliament '] = pd.to_numeric(df['Women Share of Seats in Parliament '], errors='coerce')

# Convert the 'Female Population with Some Secondary Education' column to numeric
df['Female Population with Some Secondary Education'] = pd.to_numeric(df['Female Population with Some Secondary Education'], errors='coerce')

# Convert the 'Male Population with Some Secondary Education' column to numeric
df['Male Population with Some Secondary Education'] = pd.to_numeric(df['Male Population with Some Secondary Education'], errors='coerce')

# Convert the 'Female Labour Force Participation Rate' column to numeric
df[' Female Labour Force Participation Rate'] = pd.to_numeric(df[' Female Labour Force Participation Rate'], errors='coerce')

# Convert the 'Male Labour Force Participation Rate' column to numeric
df['Male Labour Force Participation Rate'] = pd.to_numeric(df['Male Labour Force Participation Rate'], errors='coerce')

# convert Maternal Mortality Ratio to numeric
df['Maternal Mortality Ratio'] = pd.to_numeric(df['Maternal Mortality Ratio'], errors='coerce')

#convert Country column to string
df['Country'] = df['Country'].astype(str)

# Print the data types of the columns to confirm that the conversion worked
print(df.dtypes)
```

```python
#rename columns
df.rename({"Women Share of Seats in Parliament " : "Women Share of Seats in Parliament", " Female Labour Force Participation Rate" : "Female Labour Force Participation Rate"}, axis=1, inplace=True)
```

```python
df.to_csv('../data/clean_data/final.csv', index=False)
```
